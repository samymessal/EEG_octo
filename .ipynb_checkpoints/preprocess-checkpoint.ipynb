### Detrend data

Removing linear trends or slow drifts from the EEG signal, often caused by low-frequency noise or baseline shifts

```python
detrend_data = detrend(raw_data)

info = mne.create_info(ch_names=['EEG0'], sfreq=250, ch_types='eeg')
detrend_raw = mne.io.RawArray(detrend_data, info)
```

### Artifact Detection

Since we are dealing with a single-channel EEG, a robust artifact identification is complicated. Most techniques use different channels as references to extract the signals corresponding to artifacts.
Nonetheless, the data will be explored to verify that the artifacts are not causing too much noise. The technique proposed is based on amplitude changes in each epoch.
Let's start by creating epochs of 30 seconds in our data:

```python
epoch_duration = 30
raw = raw.copy().crop(0, 500)
events = mne.make_fixed_length_events(raw, duration=epoch_duration)

# Creating epochs of 30 seconds
epochs = mne.Epochs(raw, 
                    events, 
                    tmin=0, 
                    tmax=epoch_duration, 
                    baseline=None, 
                    preload=True)

artifact_times = np.empty(0, dtype=int)
sfreq = raw.info['sfreq']  # Sampling frequency

for i, epoch in enumerate(epochs):
    epoch_data = epoch[0]
    artifact_indices = utils.detect_amplitude_artifacts(epoch_data, 5)
    adjusted_indices = artifact_indices + int(i * epoch_duration * sfreq)
    artifact_times = np.append(artifact_times, adjusted_indices)

times = np.arange(detrend_data.size) / raw.info['sfreq']
plt.plot(times, detrend_data[0], label='EEG')

# Plot artifact markers
plt.scatter(times[artifact_times], detrend_data[0, artifact_times], color='red', label='Artifacts', s=2)

# Plot sleep spindles if they are marked in 'labels'
# for _, row in labels.iterrows():
#     if row['SS1'] == 1:
#         plt.axvline(x=row['Timestamp'] / sfreq, color='g', label='Sleep Spindle')

plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

Filtering artifacts to only show those that are less than 30 seconds before or after a labeled event in our annotations

```python
artifact_annotation = mne.Annotations(onset=artifact_times, 
                                      duration=[0.01] * len(artifact_times),
                                      description=['Art'] * len(artifact_times))

label_annotation = utils.get_annotations(labels)

filtered_artifacts = [art for art in artifact_annotation if utils.is_artifact_close_to_labels(art, label_annotation, 30)]

filtered_art_anns = mne.Annotations(onset=[ann['onset'] for ann in filtered_artifacts],
                                       duration=[ann['duration'] for ann in filtered_artifacts],
                                       description=[ann['description'] for ann in filtered_artifacts])

filtered_annotations = filtered_art_anns + label_annotation
raw.set_annotations(filtered_annotations)
```
